```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data5 = np.loadtxt('data/1e5.txt')
data6 = np.loadtxt('data/1e6.txt')
data7 = np.loadtxt('data/1e7.txt')

plt.figure(figsize=(8, 6))
plt.title('Время выполнения от числа ядер', fontsize=16)
plt.scatter(data5[:, 0], data5[:, 1], label=r'$10^5$ чисел', zorder=10)
plt.scatter(data6[:, 0], data6[:, 1], label=r'$10^6$ чисел', zorder=10)
plt.scatter(data7[:, 0], data7[:, 1], label=r'$10^7$ чисел', zorder=10)

plt.xlabel('n, threads', fontsize=14)
plt.ylabel(r'$t, \ \mu s$', fontsize=14)
plt.legend(fontsize=14)
plt.grid()
plt.yscale('log')
```
